# **Importing libraries**

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

# **Data Collection and Analysis**

```python
df= pd.read_csv('diabetes.csv')
df.head()
```

```python
# Number of rows and columns in the data frame
df.shape
```

```python
# Getring the statisticals measures of the data
df.describe()
```

```python
df['Outcome'].value_counts()
```

0 -------> Non-Diabetic

1 -------> Diabetic

```python
df.groupby('Outcome').mean()
```

# Seperating the data and lables

```python
X = df.drop(columns='Outcome', axis=1)
```

```python
Y = df['Outcome']
```

```python
print (X)
print (Y)
```

# **Data Standarization**

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standarized_data = scaler.transform(X)
```

```python
print (standarized_data)
```

```python
X = standarized_data
Y = df['Outcome']
```

# **Train test split**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

# **Training the model**

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

# **Model Evaluation**

**Acuracy Score on training data**

```python
X_train_predection = classifier.predict(X_train)
```

```python
training_data_accurancy = accuracy_score(X_train_predection, Y_train)
```

```python
print ('The accuracy score of the training data is :',training_data_accurancy )
```

**Acuracy Score on test data**

```python
X_test_predection = classifier.predict(X_test)
```

```python
testing_data_accurancy = accuracy_score(X_test_predection, Y_test)
```

```python
print ('The accuracy score of the testing data is :',testing_data_accurancy )
```
